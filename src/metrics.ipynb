```python
import pandas as pd
from create_database import Session, Municipality, Page, Link

class MunicipalityData:
    def __init__(self):
        self.session = Session()
        self.municipalities = None
        self.pages = None
        self.links = None
        self.load_data()

    def load_data(self):
        # Query Municipality data and store it in a DataFrame
        municipalities_query = self.session.query(Municipality).all()
        municipalities_data = [(m.id, m.name, m.country, m.email, m.population, m.pop_d, m.status) for m in municipalities_query]
        self.municipalities = pd.DataFrame(municipalities_data, columns=['id', 'name', 'country', 'email', 'population', 'pop_d', 'status'])

        # Query Page data and store it in a DataFrame
        pages_query = self.session.query(Page).all()
        pages_data = [(p.id, p.url, p.content, p.municipality_id) for p in pages_query]
        self.pages = pd.DataFrame(pages_data, columns=['id', 'url', 'content', 'municipality_id'])

        # Query Link data and store it in a DataFrame
        links_query = self.session.query(Link).all()
        links_data = [(l.id, l.source_id, l.destination_id, l.depth, l.municipality_url) for l in links_query]
        self.links = pd.DataFrame(links_data, columns=['id', 'source_id', 'destination_id', 'depth', 'municipality_url'])
```

```python
mdb = MunicipalityData()
# print the first few rows of each DataFrame
print(mdb.municipalities.head())
print(mdb.pages.head())
print(mdb.links.head())
```
